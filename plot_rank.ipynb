```python
df = pd.read_csv('_'.join(reversed(files[0].split('-'))) + '.csv')
df.name = df.name.apply(lambda x: x.replace(',\xa0', ';'))
df = df.loc[:,['name', 'rank']]
cols = [files[0]]

for f in files[1:]:
    temp = pd.read_csv('_'.join(reversed(f.split('-'))) + '.csv')
    temp.name = temp.name.apply(lambda x: x.replace(',\xa0', ';'))
    temp = temp.loc[:,['name', 'rank']]
    df = pd.merge(df, temp, how='outer', on='name')
    cols.append(f)
    df.columns = ['name'] + cols
```

```python
df = df.T
df.columns = df.iloc[0,:]
df = df.drop('name')
df = df.applymap(lambda x: float(x))
df = df.drop(';', axis=1)
df.iloc[-5:,-5:]
```

```python
players_00s = ['Federer;Roger',
               'Nadal;Rafael',
               'Djokovic;Novak',
               'Murray;Andy']
players_80s = ['McEnroe;John',
               'Lendl;Ivan',
               'Wilander;Mats',
               'Connors;Jimmy']
players_90s = ['Sampras;Pete',
               'Agassi;Andre']
player = ['Dimitrov;Grigor','Nishikori;Kei','Raonic;Milos']
plot_df = df.loc[:,player]
```

```python
melted = pd.melt(plot_df.reset_index(), id_vars='index')
melted['index'] = melted['index'].apply(lambda x: pd.to_datetime(x))
ggplot(aes(x='index', y='value', colour='name'), data=melted) + \
    geom_line(linetype='-.') + \
    geom_point() + \
    scale_y_reverse() + \
    ylim(1,100)
```

```python
def get_weeks_in_top(player, rank):
    in_top = player[player <= rank].dropna()
    if len(in_top) > 0:
        return int( (pd.to_datetime(in_top.index[-1]) - pd.to_datetime(in_top.index[0])).days / 7 )
    else:
        return 0
```

```python
weeks_in_top = [get_weeks_in_top(df.loc[:,i], 10) for i in df.columns]
weeks = pd.DataFrame(list(dict(zip(df.columns, weeks_in_top)).items()), columns=['name', 'weeks'])
weeks.head()
```
